```python
folder = Path(r'G:\Мой диск\Коммерческие сметы\1. ФАКТ\разноски')

data = []
for file in folder.rglob('*.xlsx'):
    try: 
        wb = load_workbook(file, read_only=True)
        for sheet in wb.sheetnames:
            data.append({'Файл' : file.name, 'Путь' : str(file), 'Вкладка' : sheet})
        wb.close()
    except Exception as e:
        print(f'Ошибка при чтении {file.name}: {e}')

df = pd.DataFrame(data)
df.head()
```
